## MODEL BUILDING

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import LinearSVR, SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor

from sklearn.model_selection import GridSearchCV
```

we are using different regressors to know which performs the best

```python
y = df['strength'].copy()
X = df.drop('strength', axis=1)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
    
    
# Scale X with a standard scaler
scaler = StandardScaler()
scaler.fit(X_train)
    
X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import LinearSVR, SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
models = {
    "                     Linear Regression": LinearRegression(),
    "                 L2 (Ridge) Regression": Ridge(),
    "Support Vector Machine (Linear Kernel)": LinearSVR(),
    "   Support Vector Mahcine (RBF Kernel)": SVR(),
    "                         Decision Tree": DecisionTreeRegressor(),
    "                        Neural Network": MLPRegressor(),
    "                         Random Forest": RandomForestRegressor(),
    "                     Gradient Boosting": GradientBoostingRegressor(),
    "                              AdaBoost": AdaBoostRegressor()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + " trained.")
```

### Feature Importance

```python
model = GradientBoostingRegressor() # try out and change with different regressors, feature_importances will get changed
model.fit(X_train,y_train)

model.feature_importances_
```

```python
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
feature_imp = pd.Series(model.feature_importances_, df.columns[:-1]).sort_values()
feature_imp.plot(kind='barh', title='Feature Importances')
plt.xlabel('Feature Importance Score')
```

```python
for name, model in models.items():
    print(name + " R^2: {:.5f}".format(model.score(X_test, y_test)))
```

```python
best_model = GradientBoostingRegressor() 
best_model.fit(X_train, y_train)

print("Model R^2 (Before Optimization): {:.5f}".format(best_model.score(X_test, y_test)))
```

```python
params = { 'learning_rate': [0.01, 0.1, 1.0],
           'n_estimators': [100, 150, 200],
           'max_depth': [3, 4, 5] }

clf = GridSearchCV(best_model, params)
clf.fit(X_train, y_train)

clf.best_params_
```

```python
params = {  'learning_rate': [0.01, 0.1, 1.0],
            'n_estimators': [100, 150, 200],
            'max_depth': [3, 4, 5] }

clf = GridSearchCV(best_model, params)
clf.fit(X_train, y_train)

clf.best_params_
```

I found GradientBoostRegressor works more efficiently, which also provides the predominent features which performed well with the other features and with the target.

where as the GradientBoostRegressor is a boosting technique which performed the residual error perfectly to give the best score for the testing data.

Data distribution was found to be not defined perfectly because of raw and some unstructured way the data was presented. But to perform the regression, the data has been scaled and transformed for the training purpose.

strength column is been took as the target variable and the other variables as features.
